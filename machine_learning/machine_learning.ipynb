# Sales Prediction:  

For our analysis, we used stores sales historical data from Kaggle article “Retail Analysis with Walmart Data”. This data represent the sales time series of Walmart Store sales. This is the historical data that covers sales from 2010-02-05 to 2012-11-01.
Time series forecasting is one of the major building blocks of Machine Learning. For time series forecasting, approaches such as linear models, ARIMA algorithm, Random Forest, Lasso Regression, Neural Network are widely used. Machine lerning algorithm make it possible to find patterns in the time series. Sometimes we need to forecast not only more probable values of sales but also their distribution. 
Here, we will start the analysis by using the linear model for sales forecasting. After that we will do the analysis using other above mentioned machine learning models.

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
sales_df = pd.read_csv(Path('../Resources/Walmart_Store_sales.csv'))
sales_df.head(10)
```

```python
plt.scatter(sales_df.Fuel_Price, sales_df.Weekly_Sales)
plt.xlabel('Fuel_Price')
plt.ylabel('Weekly_Sales')
plt.show()
```

```python
plt.scatter(sales_df.Temperature, sales_df.Weekly_Sales)
plt.xlabel('Temperature')
plt.ylabel('Weekly_Sales')
plt.show()
```

```python
plt.scatter(sales_df.CPI, sales_df.Weekly_Sales)
plt.xlabel('CPI')
plt.ylabel('Weekly_Sales')
plt.show()
```
